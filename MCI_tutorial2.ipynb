```python
import numpy as np
from scipy.stats import shapiro
import random
import pandas as pd

dataset = np.random.normal(loc=9.5, scale=1.0, size=1000)
print(dataset)

# checking if data indeed comes from normal distribution
sh = shapiro(dataset)
print(sh)
```

```python
# T = W + gaussian_noise
mu = 0
sigma = 3

w = np.random.normal(loc=9.5, scale=1.0, size=1000)
noise1 = np.random.normal(mu, sigma, dataset.size)
t = w + noise1

t = [elem for elem in t]

# binarising the values (randomly)
binarised_pairs = []
for elem in t:
    if elem >= 9.5:
        binarised_pairs.append((1,elem))
    else:
        binarised_pairs.append((0,elem))
        
# print(binarised_pairs)

noise2 = np.random.normal(mu, sigma, dataset.size)
Y = w + t + noise2
Y1 = w + 1 + noise2
Y0 = w + 0 + noise2

# print(Y1)

# adding the differences --> E[Y1-Y0]
difference_A = 0
for i in range(len(Y)):
    difference_A += (Y1[i] - Y0[i])
print(difference_A/len(Y))

# adding the differences --> E[Y  | T=1] - E[Y | T=0]
treatment_1_lst = [pair[1] for pair in binarised_pairs if pair[0]==1]
treatment_1 = sum(treatment_1_lst)/len(treatment_1_lst)

treatment_0_lst = [pair[1] for pair in binarised_pairs if pair[0]==0]
treatment_0 = sum(treatment_0_lst)/len(treatment_0_lst)

difference_B = treatment_1 - treatment_0
print(difference_B)
```
