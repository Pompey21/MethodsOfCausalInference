```python
import numpy as np
from scipy.stats import shapiro
import random
import pandas as pd

dataset = np.random.normal(loc=9.5, scale=1.0, size=1000)
print(dataset)

# checking if data indeed comes from normal distribution
sh = shapiro(dataset)
print(sh)
```

```python
# T = W + gaussian_noise
mu = 0
sigma = 3

w = np.random.normal(loc=9.5, scale=1.0, size=1000)
noise1 = np.random.normal(mu, sigma, dataset.size)
t = w + noise1

t = [elem for elem in t]

# binarising the values (randomly)
binarised_pairs = []
for elem in t:
    if elem >= 9.5:
        binarised_pairs.append((1,elem))
    else:
        binarised_pairs.append((0,elem))
        
# print(binarised_pairs)

noise2 = np.random.normal(mu, sigma, dataset.size)
Y = w + t + noise2
Y1 = w + 1 + noise2
Y0 = w + 0 + noise2

# print(Y1)

# adding the differences
difference = 0
for i in range(len(Y)):
    difference += (Y1[i] - Y0[i])
print(difference/len(Y))
```

```python
def corr_vars( start=5, stop=10, step=0.5, mu=0, sigma=3, func=lambda x: x ):
    # Generate x
#     x = np.arange(start, stop, step)  
    x = np.random.normal(loc=9.5, scale=4.0, size=1000)
     
    # Generate random noise
    w = np.random.normal(mu, sigma, x.size)
     
    # Generate y values as y = func(x) + e
    y = np.zeros(x.size)
     
    for ind in range(x.size):
        y[ind] = func(x[ind]) + w[ind]
     
    return (x,y)
```

```python
(x,y) = corr_vars()
```

```python
print((x,y))
```

```python
# generating a pair for each of the points - COUNTER FACTUALS
def generate_CF(one_x):
    return [elem for elem in np.random.normal(loc=one_x, scale=1.0, size=1)]
```

```python
cf = generate_CF(8.1819)
print(cf)

observed = [elem for elem in y]
cfs = [item for sublist in [generate_CF(elem) for elem in observed] for item in sublist]
print(cfs)

print()
print()
print(observed)
```

```python
# need to. binarise the values
def binarisation():
    return random.randint(0,1)
binary_col = [binarisation() for elem in observed]
print(binary_col)
```

```python
df = pd.DataFrame(
    {'treatment': binary_col,
     'observed': observed,
     'cfs': cfs
    })
print(df)
```

```python
def rearrange(df):
    for index in range(0,1000):
        if not df.iloc[index].treatment:
            df.at[index,'observed'], df.at[index,'cfs'] = max(df.iloc[index].observed, df.iloc[index].cfs), min(df.iloc[index].observed, df.iloc[index].cfs)
    
            
print(df.head())
rearrange(df)
df.head()
```

```python
# summing across the average
mean = df.mean(axis=0)
print(mean)
e_avearge = mean.observed - mean.cfs
print(e_avearge)
```
